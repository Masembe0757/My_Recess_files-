```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mt
import seaborn as sns
import sklearn as sk
```

```python
data = pd.read_csv("heart_disease.csv")
data
```

```python
#Bar plot showing heart disease frequency of age
pd.crosstab(data.age,data.target).plot(kind = 'bar',figsize = (20,6))
plt.title('Heart diseases frequency against age')
plt.ylabel('age')
plt.xlabel('age')
plt.show()
```

```python
#Bar plot showing heart disease frequency of sex
pd.crosstab(data.sex,data.target).plot(kind = 'bar',figsize = (20,6))
plt.xticks(rotation = 0)
plt.ylabel('frequency')
plt.xlabel('sex')
plt.title('Heart diseases frequency against sex')
plt.show()
```
