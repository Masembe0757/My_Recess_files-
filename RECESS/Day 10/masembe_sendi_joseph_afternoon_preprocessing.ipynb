No.5 A Pandas program to display the most frequent value in the given series and replace everything else as 'Other' in the series

```python
df = pd.read_csv("AfricaCupofNationsMatches.csv")  #loading the dataset
home_Team_goals_series = pd.Series(df["HomeTeamGoals"])  #converting target column to series
frequencies = home_Team_goals_series.value_counts() #returning frequencies of values
frequencies
```

```python
#Replacing other values with other otherthan the most frequent value "1" now that its known
for x in home_Team_goals_series:
    if x!=1:
        x="Other"
    else:
        x= 1
    print(x)
```

Another approach is to use a formulae
Here," is in " checks if the current member belongs to the index of the most frequent member and the tilde "~" negates this condition thus only members that are not the most frequent value are assigned a string Other to bring same result as above

```python
home_Team_goals_series[~home_Team_goals_series.isin(home_Team_goals_series.value_counts().index[:1])]="Other"
home_Team_goals_series.head(15)
```
