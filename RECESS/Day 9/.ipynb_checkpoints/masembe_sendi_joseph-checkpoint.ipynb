# Outliers using quartiles

```python
lower_limit = df4["Age"].quantile(0.05) #5%
lower_limit
```

```python
upper_limit = df4["Age"].quantile(0.95) #95%
upper_limit
```

```python
#Displaying records below lower limit
df4[df4["Age"]<lower_limit]
```

```python
#Displaying records above uppper limit
df4[df4["Age"]>upper_limit]
```

```python
#EXcluding outliers#Displaying records below lower limit

df4_without_outliers = df4[(df4["Age"]>lower_limit) & (df4["Age"]< upper_limit)]
df4_without_outliers
```
